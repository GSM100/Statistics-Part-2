1. What is hypothesis testing in statistics?

 - Hypothesis testing is a statistical method used to make decisions about a population based on a sample. It involves formulating two competing hypotheses and using sample data to determine which hypothesis is more likely to be true.

2. What is the null hypothesis, and how does it differ from the alternative hypothesis?

    - Null Hypothesis (H₀): A default assumption that there is no effect or no difference.

     Alternative Hypothesis (H₁ or Ha): Suggests that there is an effect or a difference.

    Example:

       H₀: μ = 100 (mean is 100)

       H₁: μ ≠ 100 (mean is not 100)


3. What is the significance level in hypothesis testing, and why is it important?

   - The significance level (α) is the probability of rejecting the null hypothesis when it is actually true. Common values: 0.05, 0.01.

    It defines the threshold for accepting or rejecting H₀ and controls the risk of a Type I error.


4. What does a P-value represent in hypothesis testing?

  - A P-value is the probability of obtaining a result at least as extreme as the observed data, assuming that the null hypothesis is true.


5. How do you interpret the P-value in hypothesis testing?

  - If P-value ≤ α: Reject H₀ (evidence supports H₁)

    If P-value > α: Fail to reject H₀ (not enough evidence)

     Smaller P-values indicate stronger evidence against the null hypothesis.

6. What are Type I and Type II errors in hypothesis testing?

   - Type I Error (False Positive): Rejecting H₀ when it is true (controlled by α)

    Type II Error (False Negative): Failing to reject H₀ when it is false (controlled by β)

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

  - One-tailed test: Tests for a directional effect (e.g., μ > 100)

    Two-tailed test: Tests for any difference (e.g., μ ≠ 100)

8. What is the Z-test, and when is it used in hypothesis testing?
A Z-test is used when:

  - The population standard deviation (σ) is known.

   The sample size is large (n ≥ 30).
It tests hypotheses about population means or proportions.

9. How do you calculate the Z-score, and what does it represent in hypothesis testing?

  - Z = (X̄ - μ) / (σ / √n)

     X̄ = sample mean

    μ = population mean

    σ = population std dev

    n = sample size

   It tells how many standard deviations the sample mean is from the population mean.

10. What is the T-distribution, and when should it be used instead of the normal distribution?

   - The T-distribution is used when:

     The population standard deviation is unknown

     The sample size is small (n < 30)

     It is wider than the normal distribution to account for extra uncertainty.


11. What is the difference between a Z-test and a T-test?

    Criteria- Population Std Dev,  	Z-test- Known , T-test-  Unknown

 Criteria- Sample Size	 ,  Z-test- Large (n ≥ 30), T-test- Small (n < 30)

 Criteria- Distribution Used, Z-test- Normal distribution, T-test- T-distribution

12. What is the T-test, and how is it used in hypothesis testing?


   - A T-test compares sample means to determine if they come from populations with the same mean. Types include:

   One-sample t-test

    Two-sample t-test (independent)

    Paired sample t-test

13. What is the relationship between Z-test and T-test in hypothesis testing?

 -  Both test hypotheses about means. The Z-test assumes known standard deviation, while the T-test does not. As sample size increases, the T-distribution approximates the normal (Z) distribution.

14. What is a confidence interval, and how is it used to interpret statistical results?

  -   A confidence interval (CI) is a range of values likely to contain the population parameter with a certain level of confidence (e.g., 95%).

   If a CI for a mean difference does not include 0, it suggests a statistically significant difference.

15. What is the margin of error, and how does it affect the confidence interval?

  - Margin of error is the amount added/subtracted from the sample estimate to form the CI. A larger margin means a wider CI, which increases confidence but reduces precision.

16. How is Bayes' Theorem used in statistics, and what is its significance?

  - Bayes’ Theorem calculates the probability of an event based on prior knowledge of related events. It's key in Bayesian statistics, used for updating beliefs with new data.

  Formula:
P(A|B) = [P(B|A) * P(A)] / P(B)

17. What is the Chi-square distribution, and when is it used?

   - A Chi-square distribution is used to assess relationships between categorical variables or test variance. It’s always positive and skewed right.

18. What is the Chi-square goodness of fit test, and how is it applied?

  - The Chi-square goodness of fit test checks if observed frequencies match expected frequencies.

   Formula:
χ² = Σ[(O - E)² / E]

  Where O = observed, E = expected

19. What is the F-distribution, and when is it used in hypothesis testing?

  - The F-distribution is used to compare variances. It's commonly used in:

   ANOVA

    F-test for variance equality

     It is right-skewed and non-negative.

20. What is an ANOVA test, and what are its assumptions?

  - ANOVA (Analysis of Variance) tests whether three or more group means are significantly different.

     Assumptions:

     Independence of observations

     Normally distributed groups

     Equal variances (homogeneity)

21. What are the different types of ANOVA tests?

    - One-way ANOVA: One independent variable

     Two-way ANOVA: Two independent variables

     Repeated Measures ANOVA: Same subjects tested multiple times

22. What is the F-test, and how does it relate to hypothesis testing?

    - The F-test compares two variances. It’s used in:

    ANOVA (comparing group means)

    Regression analysis (model significance)

    Formula:
    F = Variance₁ / Variance₂


```python
# Import necessary libraries
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

# 1. Perform a Z-test comparing sample mean to known population mean
def z_test(sample, pop_mean, pop_std):
    n = len(sample)
    sample_mean = np.mean(sample)
    z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))
    return z_score, p_value

```

```python
# 2. Simulate data and perform Z-test
np.random.seed(0)
simulated_sample = np.random.normal(loc=52, scale=10, size=50)
pop_mean = 50
pop_std = 10
z_score, p_val = z_test(simulated_sample, pop_mean, pop_std)
```

```python
 #4. Two-tailed Z-test visualization
def plot_z_test(z_score, alpha=0.05):
    x = np.linspace(-4, 4, 1000)
    y = stats.norm.pdf(x)
    z_crit = stats.norm.ppf(1 - alpha/2)

    plt.figure(figsize=(10, 5))
    plt.plot(x, y, label="Standard Normal Distribution")
    plt.fill_between(x, y, where=(x < -z_crit) | (x > z_crit), color='red', alpha=0.3, label='Rejection Region')
    plt.axvline(z_score, color='blue', linestyle='--', label=f'Z-score = {z_score:.2f}')
    plt.title("Two-tailed Z-Test Decision Region")
    plt.xlabel("Z")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()

plot_z_test(z_score)
```

```python
# 5. Visualize Type I and Type II errors
def visualize_errors(mu0=50, mu1=52, sigma=10, n=30, alpha=0.05):
    x = np.linspace(40, 60, 1000)
    se = sigma / np.sqrt(n)
    crit_val = stats.norm.ppf(1 - alpha/2)
    lower_crit = mu0 - crit_val * se
    upper_crit = mu0 + crit_val * se

    y_null = stats.norm.pdf(x, mu0, se)
    y_alt = stats.norm.pdf(x, mu1, se)

    plt.figure(figsize=(10, 6))
    plt.plot(x, y_null, label="Null Hypothesis (H0)", color='blue')
    plt.plot(x, y_alt, label="Alternative Hypothesis (H1)", color='green')
    plt.fill_between(x, y_null, where=(x < lower_crit) | (x > upper_crit), color='blue', alpha=0.2, label="Type I Error")
    plt.fill_between(x, y_alt, where=(x >= lower_crit) & (x <= upper_crit), color='green', alpha=0.2, label="Type II Error")
    plt.axvline(lower_crit, color='black', linestyle='--')
    plt.axvline(upper_crit, color='black', linestyle='--')
    plt.title("Type I and Type II Errors")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()

visualize_errors()
```

```python
# 6. Independent T-test
def perform_independent_ttest():
    np.random.seed(1)
    group1 = np.random.normal(loc=100, scale=15, size=30)
    group2 = np.random.normal(loc=110, scale=15, size=30)
    t_stat, p_value = stats.ttest_ind(group1, group2)
    print(f"Independent T-test\nT-statistic: {t_stat:.4f}, P-value: {p_value:.4f}\n")
    return t_stat, p_value

perform_independent_ttest()
```

```python
# 7. Paired T-test and visualization
def perform_paired_ttest():
    np.random.seed(2)
    before = np.random.normal(loc=80, scale=10, size=30)
    after = before + np.random.normal(loc=5, scale=5, size=30)
    t_stat, p_value = stats.ttest_rel(before, after)

    plt.figure(figsize=(8, 5))
    plt.plot(before, label="Before", marker='o')
    plt.plot(after, label="After", marker='x')
    plt.title("Paired Sample Comparison")
    plt.legend()
    plt.grid(True)
    plt.show()

    print(f"Paired T-test\nT-statistic: {t_stat:.4f}, P-value: {p_value:.4f}\n")
    return t_stat, p_value

perform_paired_ttest()
```

```python
# 8. Compare Z-test and T-test
def simulate_z_t_comparison():
    np.random.seed(3)
    sample = np.random.normal(loc=105, scale=20, size=25)
    pop_mean = 100
    pop_std = 20
    z_score = (np.mean(sample) - pop_mean) / (pop_std / np.sqrt(len(sample)))
    z_p = 2 * (1 - stats.norm.cdf(abs(z_score)))
    t_score, t_p = stats.ttest_1samp(sample, pop_mean)
    print(f"Z-test vs T-test Comparison\nZ-score: {z_score:.4f}, P-value: {z_p:.4f}\nT-score: {t_score:.4f}, P-value: {t_p:.4f}\n")
    return z_score, z_p, t_score, t_p

simulate_z_t_comparison()
```

```python
# 9. Confidence interval function
def confidence_interval(sample, confidence=0.95):
    n = len(sample)
    mean = np.mean(sample)
    sem = stats.sem(sample)
    h = sem * stats.t.ppf((1 + confidence) / 2., n-1)
    lower, upper = mean - h, mean + h
    print(f"Confidence Interval ({int(confidence*100)}%)\nMean: {mean:.2f}, CI: [{lower:.2f}, {upper:.2f}]\n")
    return lower, upper

confidence_interval(simulated_sample)
```

```python
# 11. Bayesian inference
def bayesian_inference(prior_A, prior_B, likelihood_A, likelihood_B):
    evidence = prior_A * likelihood_A + prior_B * likelihood_B
    posterior_A = (prior_A * likelihood_A) / evidence
    print(f"Posterior Probability of A: {posterior_A:.4f}\n")
    return posterior_A

bayesian_inference(0.6, 0.4, 0.7, 0.3)
```

```python
# 12. Chi-square test for independence
from scipy.stats import chi2_contingency

def chi_square_independence():
    data = np.array([[30, 10], [20, 40]])
    chi2, p, dof, expected = chi2_contingency(data)
    print(f"Chi-square Test\nChi2: {chi2:.2f}, P-value: {p:.4f}, DOF: {dof}\nExpected Frequencies:\n{expected}\n")

chi_square_independence()
```

```python
# 13. Expected frequencies manually
def expected_frequencies(data):
    row_totals = data.sum(axis=1).reshape(-1, 1)
    col_totals = data.sum(axis=0).reshape(1, -1)
    total = data.sum()
    expected = (row_totals @ col_totals) / total
    print(f"Expected Frequencies:\n{expected}\n")
    return expected

expected_frequencies(np.array([[30, 10], [20, 40]]))
```

```python
 #14. Goodness-of-fit test
def goodness_of_fit():
    observed = [18, 22, 20]
    expected = [20, 20, 20]
    chi2, p = stats.chisquare(f_obs=observed, f_exp=expected)
    print(f"Goodness-of-Fit Test\nChi2: {chi2:.2f}, P-value: {p:.4f}\n")

goodness_of_fit()
```

```python
# 15. Simulate and visualize chi-square

def simulate_chi_square():
    x = np.linspace(0, 20, 500)
    df = 4
    plt.plot(x, stats.chi2.pdf(x, df), label=f'df={df}')
    plt.title("Chi-square Distribution")
    plt.xlabel("Value")
    plt.ylabel("Probability Density")
    plt.grid(True)
    plt.legend()
    plt.show()

simulate_chi_square()
```

```python
# 16. F-test (compare variances)
def f_test():
    np.random.seed(0)
    x1 = np.random.normal(0, 1, 20)
    x2 = np.random.normal(0, 2, 20)
    f = np.var(x1, ddof=1) / np.var(x2, ddof=1)
    dfn, dfd = len(x1)-1, len(x2)-1
    p_value = 1 - stats.f.cdf(f, dfn, dfd)
    print(f"F-Test\nF-statistic: {f:.4f}, P-value: {p_value:.4f}\n")

f_test()
```

```python
# 17. ANOVA test
from scipy.stats import f_oneway

def perform_anova():
    np.random.seed(1)
    g1 = np.random.normal(100, 10, 30)
    g2 = np.random.normal(110, 10, 30)
    g3 = np.random.normal(105, 10, 30)
    f_stat, p_val = f_oneway(g1, g2, g3)
    print(f"ANOVA Test\nF-statistic: {f_stat:.2f}, P-value: {p_val:.4f}\n")

perform_anova()
```

```python
# 18. One-way ANOVA and boxplot

def one_way_anova_plot():
    np.random.seed(2)
    df = pd.DataFrame({
        'Group': ['A']*20 + ['B']*20 + ['C']*20,
        'Score': np.concatenate([
            np.random.normal(70, 5, 20),
            np.random.normal(75, 5, 20),
            np.random.normal(80, 5, 20)])
    })
    sns.boxplot(x='Group', y='Score', data=df)
    plt.title("One-way ANOVA Boxplot")
    plt.grid(True)
    plt.show()

    f_stat, p_val = f_oneway(df[df.Group == 'A'].Score,
                              df[df.Group == 'B'].Score,
                              df[df.Group == 'C'].Score)
    print(f"One-way ANOVA\nF-statistic: {f_stat:.2f}, P-value: {p_val:.4f}\n")

one_way_anova_plot()
```

```python
# 19. Check ANOVA assumptions
def check_anova_assumptions():
    from statsmodels.stats.diagnostic import lilliefors
    from statsmodels.stats.stattools import durbin_watson
    np.random.seed(3)
    data = np.random.normal(0, 1, 100)
    stat, p = lilliefors(data)
    dw = durbin_watson(data)
    print(f"Normality Test (Lilliefors): P-value = {p:.4f}")
    print(f"Independence Test (Durbin-Watson): Statistic = {dw:.4f}\n")

check_anova_assumptions()
```
